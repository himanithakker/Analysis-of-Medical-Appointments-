```python
def check_duplicates(df1):
#check for duplicates

  duplicates = df1.duplicated()

#count the number of duplicates

  duplicates_count = duplicates.sum()

#printing if there are any duplicates in the data

  if duplicates_count > 0:
    print("There are duplicates")
  else:
    print ("No duplicates")

  return duplicates_count

duplicates_count = check_duplicates(df1)
print("Total duploicates:", duplicates_count)
```

```python
#Checking null values
def check_null_values(df1):
  null_values = df1.isnull().sum()

  if null_values.any():
    print ("Null values ")
  else:
    print ("No null values")

#returning the results containg null value
  return null_values

#Function to get the count of null values

null_Values_count = check_null_values(df1)
```

```python
#converting dtype to date time using astype


# Assuming df1 is your DataFrame
df1['ScheduledDay'] = df1['ScheduledDay'].astype('datetime64[ns]')
df1['AppointmentDay'] = df1['AppointmentDay'].astype('datetime64[ns]')

# Print the DataFrame to verify the changes
print(df1.head())
```

```python
def drop_columns(df1_rename):
  dropcolumns = df1_rename.drop(columns=['PatientId', 'AppointmentID', 'Neighbourhood'])

  return dropcolumns

#Calling the function

df1_droppedcolumns = drop_columns(df1_rename)

print("New DataFrame", df1_droppedcolumns)
```

```python
import pandas as pd

# Assuming df3 is your DataFrame
def create_bin(df1_droppedcolumns):
    bins = [0,20, 40, 60, 80]
    df1_droppedcolumns['Age_Group'] = pd.cut(df1_droppedcolumns['Age'], bins, right=False)
    return df1_droppedcolumns

# Call the create_bin function and pass the DataFrame
resulting_df = create_bin(df1_droppedcolumns)

# Print the resulting DataFrame
print(resulting_df)
```

```python
def drop_null(resulting_df):
  df_drop_NA = resulting_df.dropna()
  return df_drop_NA

#Calling the function

dropped_NA = drop_null(resulting_df)
print("New Dataframe",dropped_NA )
```
